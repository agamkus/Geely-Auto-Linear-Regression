# Step 2: Data cleaning and preparation:

### Creating new column `Company` and dropping the original column `CarName`:

```python
car['Company'] = car['CarName'].apply(lambda x : x.split()[0])
```

```python
car = car.drop('CarName', axis=1)
```

### Removing the column `car_ID` as there is no use of this column in the analysis:

```python
car = car.drop('car_ID', 1)
```

### Removing duplicates, if any:

```python
car = car.drop_duplicates()
```

```python
car['Company'].unique()
```

Here, we observe that there are multiple company names which are **misspelled**. Correcting these:

```python
car['Company'] = car['Company'].replace('maxda','mazda')
car['Company'] = car['Company'].replace('Nissan','nissan')
car['Company'] = car['Company'].replace('porcshce','porsche')
car['Company'] = car['Company'].replace('toyouta','toyota')
car['Company'] = car['Company'].replace(['vokswagen','vw'],'volkswagen')
```

Symboling is a categorical variable, but it is treated as a numeric int64 type variable. Let's convert it in categorical variable.
Here Symboling is the assigned insurance risk rating,A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe.
Let's make the categorization as below:

- -1 and -1 as **Low Risk**
- 0 and 1 as **Medium Risk**
- 2 and 3 as **High Risk**

```python
car['symboling'] = car['symboling'].map({-2: 'Low Risk',-1: 'Low Risk',0: 'Medium Risk',1: 'Medium Risk',2: 'High Risk',3:'High Risk'})
```

# Step 3: Deriving new feature:

### Deriving new feature `Range`:
`Range` indicates binning of the price as below:
- Price below 10000 : **Low**
- Price between 10000 and 20000 : **Mediam**
- Price above 20000 : **High**

```python
car['Range'] = car['price'].apply(lambda x : 'Low' if x < 10000 else ('Medium' if x > 10000 and x < 20000 else 'High'))
```

# Step 7: Model Building:

### Dividing data into X and y variables:

```python
y_train = df_train.pop("price")
X_train = df_train
```

### Importing RFE and other libraries:

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

##### Using RFE to get to 10 significant columns:

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, 10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

##### Columns supported by RFE:

```python
col = X_train.columns[rfe.support_]
```

##### Columns not supported by RFE:

```python
X_train.columns[~rfe.support_]
```

```python
col
```

### Building model using statsmodel, for the detailed statistics:

##### Importing library: 

```python
import statsmodels.api as sm
```

```python
X_train_rfe = X_train[col]
X_train_lm = sm.add_constant(X_train_rfe)
X_train_lm.head()
```

### Model: 1

```python
lm_1 = sm.OLS(y_train, X_train_lm).fit()
print(lm_1.summary())
```

##### Creating function for VIF:

```python
def show_vif():
    vif = pd.DataFrame()
    vif['Feature'] = X_train_rfe.columns
    vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by='VIF', ascending=False)
    return vif
```

```python
show_vif()
```

We can drop `enginelocation` due to very high VIF:

```python
X_train_rfe = X_train_rfe.drop('enginelocation',1)
```

### Model: 2

```python
X_train_lm = sm.add_constant(X_train_rfe)
lm_2 = sm.OLS(y_train, X_train_lm).fit()
print(lm_2.summary())
```

# Step 11: Equation of the best fit line:

**$ price = 0.4469 + 0.4932  \times  horsepower - 0.4378 \times Range\_ Low - 0.3322 \times Range\_Medium - 0.2140 \times enginetype\_dohcv $**

This equation explains how the price of the vehical changes with a unit change in any of these independent variable with all other variables held constant
